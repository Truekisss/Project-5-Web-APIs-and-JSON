```python
# Creating small subset of movie_id.
movie_id = [0, 299534, 19995, 140607, 299536, 597, 135397,
            420818, 24428, 168259, 99861, 284054, 12445,
            181808, 330457, 351286, 109445, 321612, 260513]
```

```python
# Basic URL for the movie module with two placeholders, for themovie ID and the API key
basic_url = 'https://api.themoviedb.org/3/movie/{}?{}'
```

```python
# Automating and scaling theprocess.
# Saving json data in the jason_list for all moviesby creating empty list json_list[]
# Then we itterate over movie_id.
# For each and every movie we replace th eplaceholders with movie and api_key.
# Creating response object with request to get and pass url.
# Introducing a conditional statement to check whether the status code of response up check is equal to 200.
    # 200 is the desired code, this means that the data is available.
    # If the status is unequal to 200, itteration will stop and go to the next movie ID.
    # If the movie id exists, then we get the data in json format and store it in the variable data.
    # Then we append to json data to our json list.
    # Finally converting complete json list to a data frame
json_list = []
for movie in movie_id:
    url = basic_url.format(movie, api_key)
    r = requests.get(url)
    if r.status_code != 200:
        continue
    else:
        data = r.json()
        json_list.append(data)
df = pd.DataFrame(json_list)
```

```python
# Selecting only some columns and sorting the movie by the revenue from high to low.
# Actually overwriting df.
df = df.loc[:, ["title", "id", "revenue", "genres", "belongs_to_collection", "runtime"]].sort_values(by = "revenue",
                                                                                                ascending = False)
```

```python
# Saving data set by exporting to the json file with to.json.
# Records orientation is the best type of orientation to work with pandas.
df.to_json("movies.json", orient = "records")
```

```python
# Oppening json file and loading into data.
with open("movies.json") as f:
    data = json.load(f)
```

```python
# Normalizzing and flattening dataframe.
pd.json_normalize(data)
```

```python
# Going deeper in to nested data, actually flattening and creating a new data frame for genres column. 
pd.json_normalize(data, "genres", "title")
```

As a summary if we export our dataset to a json file and then if we reimport, then the full functionality is still available here.
So we have pd.json_normalized and all of the optionalities. Seems to be the best practic eon how to save and export data frame with nested json data, by simply exporting to json file.
